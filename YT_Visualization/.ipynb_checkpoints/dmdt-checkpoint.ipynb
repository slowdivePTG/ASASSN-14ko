```python
def dmdt(DIR, chk, Period, eta=0.06, ax=None, Flux=False):
    try:
        if ax == None:
            f, ax = plt.subplots(2, 1, figsize=(10, 10))
    except:
        pass
    os.chdir('/Users/chang/Desktop/Santa Cruz/TDE_plot')
    os.chdir(DIR)

    if Period < 0:
        filename = 'b10000_dm_de_chk_{}.npz'.format(chk)
    else:
        filename = 'b10000_dm_de_chk_{}_{}.npz'.format(chk, Period)

    npzfile = np.load(filename)
    output = npzfile['x']
    output_bd = npzfile['y']
    output_mdot = npzfile['z']

    #dm_de v.s. e
    e, dm_de, s_dm_de = output[:, 0], output[:, 1], output[:, 2]
    ax[0].scatter(e, dm_de, s=0.5, label='{}, P = {}'.format(DIR, Period))
    ax[0].set_yscale('log')
    ax[0].set_xlabel('$\epsilon$ (erg/g)', fontsize=20)
    ax[0].set_ylabel(r'd$M$/d$\epsilon$ (g$\cdot$s$^2$/cm$^2$)', fontsize=20)

    #Mdot v.s. t
    t = output_mdot[:, 0]
    mdot, mdot_orig = output_mdot[:, 1], output_mdot[:, 2]
    D = 188 * u.Mpc
    F = (mdot * u.Msun / u.yr * eta * u.c**2 / np.pi / 4 /
         D**2).in_units('erg/s/cm**2')
    F_orig = (mdot_orig * u.Msun / u.yr * eta * u.c**2 / np.pi / 4 /
              D**2).in_units('erg/s/cm**2')
    if Flux:
        ax[1].scatter(t, F_orig, s=0.1, alpha=0.5)
        ax[1].scatter(t, F, s=1, label='{}, P = {}'.format(DIR, Period))
        ax[1].set_ylabel(r'$F$ (erg/s/cm$^2$)', fontsize=20)
    else:
        ax[1].scatter(t, mdot_orig, s=0.1, alpha=0.5)
        ax[1].scatter(t, mdot, s=1, label='{}, P = {}'.format(DIR, Period))
        ax[1].set_ylabel(r'$\dot M$ (Msun/yr)', fontsize=20)
    ax[1].set_xlabel('Time (day)', fontsize=20)
    plt.tight_layout()
    return ax
```

```python
ax = dmdt(DIR='m0.8_p17_b0.6_P110', chk='0060', Period=50, Flux=True)
ax = dmdt(DIR='m0.8_p17_b0.6_P110', chk='0060', Period=114, Flux=True, ax=ax)
ax = dmdt(DIR='m0.8_p17_b0.6_P110', chk='0060', Period=-1, Flux=True, ax=ax)
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[0].legend()
ax[1].legend()
```

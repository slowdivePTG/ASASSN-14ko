```python
import yt
import yt.units as u
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
from data_binning import *
plt.rcParams.update({
    "text.usetex": True})
```

```python
G = u.gravitational_constant
q0 = 1e6
beta = 1
M = q0 * u.Msun
rT = q0**(1 / 3) * u.Rsun
deltaE = (G * M / rT**2 * u.Rsun).in_cgs()
```

```python
class TDE:
    G = u.gravitational_constant
    q0 = 1e6
    beta = 1
    M = q0 * u.Msun
    rT = q0**(1 / 3) * u.Rsun
    deltaE = (G * M / rT**2 * u.Rsun).in_cgs()
    age0 = '0.445'
    ms0 = '1.0'

    def __init__(self,
                 q=q0,
                 Ecc=1,
                 beta=1,
                 file='1.000.dat',
                 ms=ms0,
                 age=age0,
                 DIR=None):
        if DIR == None:
            DIR = '../../STARS_library/retrieval/m{}_t{}/'.format(ms, age)
        with open(DIR + file) as f:
            lines = (line for line in f if not line.startswith('"'))
            dmdt_t0 = np.loadtxt(lines, skiprows=1)
        t0 = dmdt_t0[:, 0]
        dmdt0 = dmdt_t0[:, 1]
        dedt0 = (2 / 3 * (G**2 * M**2 * np.pi**2 / 2)**(1 / 3) *
                 (t0 * u.day)**(-5 / 3)).in_cgs()
        dmde0 = (dmdt0 * (u.Msun / u.yr) / dedt0).in_cgs()
        E = -((G**2 * M**2 * np.pi**2 / 2)**(1 / 3) *
              (t0 * u.day)**(-2 / 3)).in_cgs()
        E_dE = E / deltaE
        dmde_dE = dmde0 * deltaE
        self.M = q * u.Msun
        self.Ecc = Ecc
        self.beta = beta
        self.rT = q**(1 / 3) * u.Rsun
        self.deltaE = (G * self.M / self.rT**2 * u.Rsun).in_cgs()

        self.E = self.E_Ecc(E_dE * self.deltaE)
        self.E_most_bound = self.E_Ecc(-self.deltaE * self.beta**2)
        self.T = ((-self.E**3 * 2)**(-1 / 2) * np.pi * G *
                  self.M).in_units('day')
        self.Tfallback = self.T[0]

        self.dmde = dmde_dE / self.deltaE
        dedt = 2 / 3 * ((G**2 * self.M**2 * np.pi**2 / 2)**(1 / 3) *
                        self.T**(-5 / 3)).in_cgs()
        self.dmdt = (dedt * self.dmde).in_units('Msun/yr')

        self.Tpeak = self.T[np.argmax(self.dmdt)].in_units('day')
        self.Tpeak0 = self.Tpeak - self.Tfallback

        self.Tscale = (self.T - self.Tpeak) / (self.Tpeak - self.T[0])
        self.dmdtscale = self.dmdt / self.dmdt.max()
        self.Tscale *= 1 - self.dmdtscale[0]

        if self.Eorb == 0:
            self.Period = np.inf
        else:
            self.Period = 2 * np.pi * G * self.M / (-2 * self.Eorb)**(3 / 2)
            self.Period = self.Period.in_units('day')

    def E_Ecc(self, E):
        if self.Ecc == 1:
            self.Eorb = 0
        else:
            a = self.rT / self.beta / (1 - self.Ecc)
            self.Eorb = -G * self.M / 2 / a
        return E + self.Eorb
```

```python
def P2Ecc(P, M=1 * u.Msun, R=1 * u.Rsun, beta=1):
    return 1 - (4 * np.pi**2 / P**2 / G / M)**(1 / 3) * R / beta
```

```python
ecc = P2Ecc(114.2*u.day)
tde = TDE(Ecc=ecc, file='1.000.dat', q=2e7)
plt.figure(figsize=(8, 8))
plt.plot(tde.T-tde.Tpeak, tde.dmdt)
plt.gca().tick_params(labelsize=25)
plt.xlabel('$T-T_\mathrm{peak}$', fontsize=25)
plt.ylabel('$\dot M$', fontsize=25)
```

```python
T0 = 3000


def T(l):
    dmdt = tde.dmdt / tde.dmdt.max()
    arg = np.argwhere(dmdt > 0).flatten()
    res = dmdt**(1 / 4)
    res[arg] **= (1 - 2 * l)
    return 20000 * (res)


def Fnu(T0, l):
    return tde.dmdt * T0**3 / (np.exp(T0 / T(l)) - 1) / T(l)**4


plt.figure(figsize=(8, 8))

plt.plot(tde.T - tde.Tpeak,
         Fnu(T0, 0.5) / Fnu(T0, 0.5).max(),
         color='crimson',
         alpha=1)
plt.plot(tde.T - tde.Tpeak,
         Fnu(T0, 0) / Fnu(T0, 0).max(),
         color='crimson',
         alpha=0.8)
plt.plot(tde.T - tde.Tpeak,
         Fnu(T0, 1) / Fnu(T0, 1).max(),
         color='crimson',
         alpha=0.4)

plt.plot(tde.T - tde.Tpeak,
         Fnu(T0 * 2, 0.5) / Fnu(T0, 0.5).max(),
         color='green',
         alpha=1)
plt.plot(tde.T - tde.Tpeak,
         Fnu(T0 * 2, 0) / Fnu(T0, 0).max(),
         color='green',
         alpha=0.8)
plt.plot(tde.T - tde.Tpeak,
         Fnu(T0 * 2, 1) / Fnu(T0, 1).max(),
         color='green',
         alpha=0.4)

plt.plot(tde.T - tde.Tpeak,
         Fnu(T0 * 10, 0.5) / Fnu(T0, 0.5).max(),
         color='purple',
         alpha=1)
plt.plot(tde.T - tde.Tpeak,
         Fnu(T0 * 10, 0) / Fnu(T0, 0).max(),
         color='purple',
         alpha=0.8)
plt.plot(tde.T - tde.Tpeak,
         Fnu(T0 * 10, 1) / Fnu(T0, 1).max(),
         color='purple',
         alpha=0.4)

plt.plot(tde.T - tde.Tpeak, tde.dmdt / tde.dmdt.max(), color='k', linewidth=5)

plt.gca().tick_params(labelsize=25)
plt.xlabel('$T-T_\mathrm{peak}$', fontsize=25)
plt.ylabel('$\dot M$', fontsize=25)

plt.yscale('log')
plt.ylim(1e-3, 1e2)
```
